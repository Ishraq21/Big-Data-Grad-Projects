```python
trackingDf = pd.read_csv('tracking2020.csv')
```

```python
trackingDf.head()
```

```python
nflId_gameId_Df = trackingDf.groupby(["nflId",'gameId'])
```

```python
# Calculating average speed in yards/second for 2020

calculateSpeedMean = nflId_gameId_Df[['s']].mean().groupby('nflId').mean()
calculateSpeedMean.head()

```

```python
# Merge Average Speed
playerDf = pd.merge(playerDf,calculateSpeedMean,on=['nflId'])
```

```python
playerDf = playerDf.rename(columns={'s': 'averageSpeed2020'})
```

```python
newDf = playerDf.loc[:,~playerDf.columns.duplicated()]
newDf.head()
```

```python
# Average distance traveled per game, in yards

averageDistance = nflId_gameId_Df[['dis']].mean().groupby('nflId').mean()
averageDistance.head()
```

```python
# Merge Average Distance
newDf = pd.merge(newDf,averageDistance,on=['nflId'])
```

```python
# Renaming Tracking S to averageDistance2020

newDf = newDf.rename(columns={'dis': 'averageDistance2020'})
newDf = newDf.loc[:,~newDf.columns.duplicated()]
newDf.head()
```

# Artificial Neural Network for Prediction

## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import load_model
```

```python
# New DataFrame with Name, Position, Weight, AverageSpeed2020, AverageDistance2020 for Regression

regDf = newDf[['displayName','Position','weight','averageSpeed2020','averageDistance2020']]
regDf
```

```python
# Drop Punters, Kickers

pos = ['P','K']
regDf = regDf[regDf.Position.isin(pos) == False]
```

```python
regDf.describe()
```

```python
sns.pairplot(regDf)
```

```python
regDf.plot(x='averageSpeed2020', y='averageDistance2020', style='o',figsize=(20,20))
plt.xlabel('averageSpeed2020')
plt.ylabel('averageDistance2020')
plt.title('Average Speed vs Average Distance')
plt.show()
```

```python
# Convert into NumPy arrays

X = regDf[['weight','averageSpeed2020']].values
y = regDf['averageDistance2020'].values
```

```python
# Test/Train split using sklearn.model_selection
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
model = Sequential([Dense(4,activation='relu'),Dense(2,activation='relu'),Dense(1)])
```

```python
model = Sequential()
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

model.add(Dense(1))
```

```python
model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train, y=y_train, epochs=250)
```

```python
lossDf = pd.DataFrame(model.history.history)
lossDf.head()
```

```python
lossDf.plot(figsize=(20,20))
```

## Model Evaluation

```python
model.evaluate(X_test,y_test,verbose=0) # returns loss or mean
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions = model.predict(X_test)
test_predictions
```

```python
# Converting our predictions into Pandas Series
test_predictions = pd.Series(test_predictions.reshape(539,))
test_predictions
```

```python
# Convert the series into dataframe

predictDf = pd.DataFrame(y_test,columns=['averageDistanceFromTestData'])
predictDf = pd.concat([predictDf, test_predictions], axis=1)
predictDf.columns = ['AverageDistanceFromTestData','predictedDistance']
predictDf
```

```python
sns.scatterplot(x='AverageDistanceFromTestData',y='predictedDistance', data=predictDf)
```

```python
# Calculate the mean absolute error to see on average how far we are off from our actual price point.
mean_absolute_error(predictDf['AverageDistanceFromTestData'], predictDf['predictedDistance'])
```

```python
mean_squared_error(predictDf['AverageDistanceFromTestData'], predictDf['predictedDistance'])
```
